# Establishing a proxy for SiO2 in subalkaline volcanic rocks

The goal here is to establish a proxy for SiO2 in subalkaline volcanic rocks. The proxy needs to be unaffected by alteration (use of immobile elements ratios or ALR). The workflow is the following:

1) Opening the dataset, doing QA/QC

2) Doing a series of conventional geochemical plots to display the data

3) Testing linear regression models (Ordinary Least Squares, Ridge Regression and Lasso)

4) Testing linear regression models with engineered polynomial features to account for non-linear relations

5) Testing SVM, RF and GTB algorithms. These are non-linear, but do not provide a final 'equation' for the regression.

## 3) Linear regression

The workflow is the following:

a) creation of X and y

b) creation of training and testing sets

c) creation of the algos and a grid search (only Lasso for the moment)

d) launching the grid search and selecting the best parameters

e) estimating success scores on test data

```python
def make_ALR(dataframe, element):
    temp_df = pd.DataFrame()
    for index, Series in dataframe.iterrows():
        Series = Series.apply(lambda x: x/Series[element])
        Series = np.log(Series)
        temp_df = temp_df.append(Series)
        temp_df.drop(element, axis=1, inplace=True)
    return temp_df
```

#### a) Creating X and y

```python
#### creating x and y
X = df.loc[:, ['TIO2(WT%)', 'AL2O3(WT%)', 'Y(PPM)', 'ZR(PPM)',
               'NB(PPM)', 'LA(PPM)', 'CE(PPM)', 'SM(PPM)', 'YB(PPM)', 'TH(PPM)']]
X = make_ALR(X, 'YB(PPM)')
y = df['SIO2(WT%)']
```

#### b) creating train and test datasets

```python
#### creating test and train datasets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

#### c and d) creating the algos and doing a cross-validation to get the best parameters

```python
from sklearn.linear_model import LassoCV, Lasso

##############################################################################
# LassoCV: coordinate descent

# Compute paths
print("Computing regularization path using the coordinate descent lasso...")
model = LassoCV(cv=20).fit(X_train, y_train)

# Display results
m_log_alphas = -np.log10(model.alphas_)

print ('alpha: CV estimate: %s' % str(model.alpha_))
print ('R2 score is:')
print (model.score(X_train,y_train))
text = 'The linear equation is: SiO2 = '
for i in range(len(X.columns.values)):
    text += 'log('
    text += str(X.columns.values[i])
    text += '/YB(PPM))*'
    text += str(model.coef_[i])
    text += ' + '
print (text)

plt.figure()
ymin, ymax = 0, 150
plt.plot(m_log_alphas, model.mse_path_, ':')
plt.plot(m_log_alphas, model.mse_path_.mean(axis=-1), 'k', label='Average across the folds', linewidth=2)
plt.axvline(-np.log10(model.alpha_), linestyle='--', color='k', label='alpha: CV estimate')

plt.legend()

plt.xlabel('-log(alpha)')
plt.ylabel('Mean square error')
plt.title('Mean square error on each fold')
plt.axis('tight')
plt.ylim(ymin, ymax)
plt.show()
```

#### e) Estimate success score

```python
y_pred = model.predict(X_test)
R2 = r2_score(y_test, y_pred)
print ('The r2 score obtained on the test dataset is: ' + str(R2))
```

```python
y_pred = model.predict(X_test)

from scipy.stats.stats import pearsonr
corr = pearsonr(y_test, y_pred)

fig, ax = plt.subplots()
line = ax.plot([30, 90], [30, 90], 'r', linewidth=4, ls='--')
ax.scatter(y_test, y_pred)
ax.set(xlim=[30, 90], ylim=[30, 90])
ax.set(title='Prediction vs analysis plot with the 1:1 line')
ax.set_xlabel('SiO2 analysis (wt%)')
ax.set_ylabel('SiO2 predicted (wt%)')
ax.text(55, 40, 'Pearson correlation coeficient: %.2f' % (corr[0]))
plt.show()
```

## 4) Linear regression with polynomial features

#### a) Creating x and y

```python
#### creating x and y
X = df.loc[:, ['TIO2(WT%)', 'AL2O3(WT%)', 'Y(PPM)', 'ZR(PPM)',
               'NB(PPM)', 'LA(PPM)', 'CE(PPM)', 'SM(PPM)', 'YB(PPM)', 'TH(PPM)']]
X = make_ALR(X, 'YB(PPM)')
y = df['SIO2(WT%)']
```

#### b) creating polynomial features and testing

Polynomial features of degree 2 to 5 are created and tested using the same workflow as above. These new features allow to get complex, non-linear interactions between features.

```python
from sklearn.preprocessing import PolynomialFeatures
list_Xs = {}
score_Xs = []
for i in range(2,5):
    ####### Create polynomial features of degree i
    poly = PolynomialFeatures(i)
    list_Xs['X'+str(i)] = poly.fit_transform(X)
    
    ####### Create training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(list_Xs['X'+str(i)], y, test_size=0.33, random_state=42)
    
    ####### Compute paths from LassoCV
    print("Computing regularization path using the coordinate descent lasso on dataset X" + str(i))
    model = LassoCV(max_iter=10000, cv=20, n_jobs=3).fit(X_train, y_train)

    ####### Display results
    m_log_alphas = -np.log10(model.alphas_)

    print ('alpha: CV estimate: %s' % str(model.alpha_))
    print ('R2 score is:')
    print (model.score(X_train,y_train))
    
    ####### Calculate scores on testing set
    y_pred = model.predict(X_test)
    R2 = r2_score(y_test, y_pred)
    score_Xs.append(model.score(X_train,y_train))
    print ('The r2 score obtained on the test dataset is: ' + str(R2))
    
    ###### Display correlation on testing set
    y_pred = model.predict(X_test)
    corr = pearsonr(y_test, y_pred)
    
    fig, ax = plt.subplots()
    line = ax.plot([30, 90], [30, 90], 'r', linewidth=4, ls='--')
    ax.scatter(y_test, y_pred)
    ax.set(xlim=[30, 90], ylim=[30, 90])
    ax.set(title='Prediction vs analysis plot with the 1:1 line')
    ax.set_xlabel('SiO2 analysis (wt%)')
    ax.set_ylabel('SiO2 predicted (wt%)')
    ax.text(55, 40, 'Pearson correlation coeficient: %.2f' % (corr[0]))
    plt.show()
    
    print('\n', '\n', '\n', '\n', '\n', '\n', '\n')
```

## 5) Testing non-linear regression

#### a) creating x and y

```python
#### creating x and y
X = df.loc[:, ['TIO2(WT%)', 'AL2O3(WT%)', 'Y(PPM)', 'ZR(PPM)',
               'NB(PPM)', 'LA(PPM)', 'CE(PPM)', 'SM(PPM)', 'YB(PPM)', 'TH(PPM)']]
X = make_ALR(X, 'YB(PPM)')
y = df['SIO2(WT%)']
```

```python
#### creating test and train datasets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

#### b) Testing SVM

```python
C_range = 10.0 ** np.arange(-1, 7)
gamma_range= 10.0 ** np.arange(-5, 1)
param_grid = dict(svr__C=C_range, svr__gamma=gamma_range)
SVM = make_pipeline(StandardScaler(), SVR(cache_size=2000))
# scoring options: ['accuracy', 'adjusted_rand_score', 'average_precision', 'f1', 'log_loss', 'mean_squared_error', 'precision', 'r2', 'recall', 'roc_auc'] 
scores = ['r2']
print('Heat map:')
for score in scores:
    print("# Tuning hyper-parameters for %s" % score)
    print("Best parameters set found on development set:")
    grid = GridSearchCV(SVM, param_grid, cv=4, scoring='%s' % score, n_jobs=-1)
    #clf = grid.get_params()
    clf = grid.fit(X_train, y_train)
    print(clf.cv_results_['params'][clf.best_index_])
    print("Score: %0.03f" % clf.best_score_)
    print("")
    #for params, mean_score, scores in clf.grid_scores_:
         #print("%0.3f (+/-%0.03f) for %r" % (mean_score, scores.std() * 2, params))
    # Do a graphic representation
    #scores = [x[1] for x in clf.grid_scores_]
    #scores = np.array(scores).reshape(len(C_range), len(gamma_range))
```

```python
from matplotlib.colors import Normalize


class MidpointNormalize(Normalize):

    def __init__(self, vmin=None, vmax=None, midpoint=None, clip=False):
        self.midpoint = midpoint
        Normalize.__init__(self, vmin, vmax, clip)

    def __call__(self, value, clip=None):
        x, y = [self.vmin, self.midpoint, self.vmax], [0, 0.5, 1]
        return np.ma.masked_array(np.interp(value, x, y))
```

```python
print('Heat map:')
plt.figure(figsize=(8, 6))
plt.subplots_adjust(left=.3, right=0.8, bottom=0.3, top=0.8)
plt.imshow(scores, interpolation='nearest', cmap=plt.cm.hot, norm=MidpointNormalize(vmin=0.4, midpoint=0.83))
plt.xlabel('gamma')
plt.ylabel('C')
plt.colorbar()
plt.xticks(np.arange(len(gamma_range)), gamma_range, rotation=45)
plt.yticks(np.arange(len(C_range)), C_range)
plt.title('Validation accuracy')
plt.show()
```

Score metrics on the test set:

```python
####### Score on testing set
SVM = make_pipeline(StandardScaler(), SVR(gamma=0.1, C=100, cache_size=2000))
SVM.fit(X_train, y_train)
y_pred = SVM.predict(X_test)
R2 = r2_score(y_test, y_pred)
Mean_error = mean_absolute_error(y_test, y_pred)
Median_error = median_absolute_error(y_test, y_pred)
Squared_error = mean_squared_error(y_test, y_pred)
print ('Scores obtained on the test dataset:\n R2: ' + str(R2) +
       '\n Mean absolute error:' + str(Mean_error) +
       '\n Median absolute error: ' + str(Median_error) + 
      '\n Mean squared error: ' + str(Squared_error))
```

Correlation of prediction vs analysis on the test set

```python
from scipy.stats.stats import pearsonr
corr = pearsonr(y_test, y_pred)

fig, ax = plt.subplots()
line = ax.plot([30, 90], [30, 90], 'r', linewidth=4, ls='--')
ax.scatter(y_test, y_pred)
ax.set(xlim=[30, 90], ylim=[30, 90])
ax.set(title='Prediction vs analysis plot with the 1:1 line')
ax.set_xlabel('SiO2 analysis (wt%)')
ax.set_ylabel('SiO2 predicted (wt%)')
ax.text(55, 40, 'Pearson correlation coeficient: %.2f' % (corr[0]))
plt.show()
```

Correlation of the prediction vs analysis on all data using cross_val_predict to avoid overfitting

```python
from sklearn.model_selection import cross_val_predict, KFold
cv = KFold(n_splits=10, shuffle=True)
y_pred = cross_val_predict(SVM, X, y, cv=cv, n_jobs=3)


from scipy.stats.stats import pearsonr
corr = pearsonr(y, y_pred)

fig, ax = plt.subplots()
line = ax.plot([30, 90], [30, 90], 'r', linewidth=4, ls='--')
ax.scatter(y, y_pred)
ax.set(xlim=[30, 90], ylim=[30, 90])
ax.set(title='Prediction vs analysis plot with the 1:1 line')
ax.set_xlabel('SiO2 analysis (wt%)')
ax.set_ylabel('SiO2 predicted (wt%)')
ax.text(55, 40, 'Pearson correlation coeficient: %.2f' % (corr[0]))
plt.show()
```
